## Loading all necessary dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from tensorflow.keras.datasets import cifar10 
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import classification_report, confusion_matrix
```

## Loading the dataset

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
X_train.shape
```

```python
X_train[0].shape
```

```python
plt.imshow(X_train[0])
```

## Normalizing the training data

```python
X_train = X_train / 255
X_test = X_test / 255 
```

## Performing one hot encoding on test data

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

## Building and training the model

```python
model = Sequential()
```

```python
model.add(Conv2D(filters=32, kernel_size=(4, 4), input_shape=(32, 32, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=32, kernel_size=(4, 4), input_shape=(32, 32, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(256, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=2)
```

```python
model.fit(x=X_train, y=y_train, epochs=15, validation_data=(X_test, y_test), callbacks=[early_stop])
```

## Visualizing loss VS accuracy on graph

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses[['loss', 'accuracy']].plot()
```

## Creating classification_report and confusion_matrix

```python
predictions = model.predict(X_test)
```

```python
predictions_classes = np.argmax(predictions, axis=1)
```

```python
y_test_classes = np.argmax(y_test, axis=1)
```

```python
print(classification_report(y_test_classes, predictions_classes))
```

```python
confusion_matrix(y_test_classes, predictions_classes)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(confusion_matrix(y_test_classes, predictions_classes), annot=True)
```
